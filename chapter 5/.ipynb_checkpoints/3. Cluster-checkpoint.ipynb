### 비지도 학습 개론

데이터는 주어졌으나 답이 없는 상태. <br>
답이 없는 상태에서의 최소한의 것들을 실행할 수 있다.

### 지도 학습 vs 비지도 학습

지도학습 : 얻고자 하는 답으로 구성된 데이터
ex(개와 고양이 - 사진과 함께 답이 들어있다.)

비지도 학습 : 답이 정해져 있지 않은 데이터에서 숨겨진 구조를 파악한다. - 비슷한 데이터끼리 모은다.
ex(개와 고양이의 사진을 10만 장씩 줄 경우 서로 비슷한 특성끼리 분류시킨다.)

### 비지도 학습

숨겨진 구조(hidden structure)
- 데이터 포인트들은 비슷한 데이터 포인트끼리 뭉쳐있다(Hard clustering)<br>
ex) 이 데이터는 고양이가 100% 강아지가 0%임을 나타낼 경우<Br>
K-Means등등 사용




- 한 개의 데이터 포인트는 숨겨진 클러스터들의 결합이다 (Soft clustring)<br>
ex) 각각의 데이터 포인트가 여러 클러스터들의 집합이다<br> 
-> 책("100% 과학이다"가 아닌 과학 60%, 판타지 35%, 역사 5%식으로 혼합 된 경우가 있으므로)<br>
GMM등등 사용

### hard clustering

목표 : 비슷한 데이터 포인트끼리 모은다

비슷하게 뭉친 클러스터 2개를 찾을 경우(k=2)<Br>
각각의 데이터별 거리를 가지고 판별한다(컴퓨터가 해당 데이터가 무엇을 의미하는지 모름) -> 사람이 보고 판단.
    
### K 결정하기

일반적으로 완벽한 방법은 없다

1. 눈으로 확인하기
K의 개수별로 눈으로 군집 구분

**고려할 것들**
***
1) 데이터의 특성
- 어떻게 만들어진 데이터인가?<br>
ex) 남성과 여성의 몸무게 - 일반적으로 남성의 몸무게가 여성의 몸무게보다 많이 나간다.<br>
- 데이터 포인트 외 다른 feature<Br>
    
2) 분석 결과로 얻고자 하는 것<Br>
- 고양이 vs 개 분류(k = 2)<br>
- 사람들의 행동 분석(k = 5)<br>
- 가격 대비 효율성 분석
