### 확률 기초

**확률이란**

- 어떤 사건이 일어날 것인지, 혹은 일어났는지에 대한 지식 혹은 믿음을 표현하는 방법


**예제1)**<br>
학생 1000명에게 강아지를 기르시나요? : 예/아니오

사건 yes : 강아지를 기른다 -> 88명<Br>
사건 no : 강아지를 기르지 않는다 -> 912명

P(yes) = 88/1000 = 0.088<br>
P(no) = 912/1000 = 0/912<br>

***
**예제 2**<br>
완벽한 주사위를 한 번 굴린다.

짝수가 나오는 사건 A: {2,4,6}<br>
훌수가 나오는 사건 B: {1,3,5}<br>

P(A) = 0.5<br>
P(B) = 0.5

합집합 = 1<br>
교집합 = 0

***
**예제 2-1**
짝수가 나오는 사건 A: {2,4,6}<br>
3보다 큰 수가 나오는 사건 B:{4,5,6}

P(A) = 3/6 = 1/2<BR>
P(B) = 3/6 = 1/2

합집합 = 4/6 <BR>
교집합 = 2/6

**조건부확률** : 사건 B가 일어났을 때 A가 일어날 확률

**3보다 큰 수가 나왔을 때 그 수가 짝수일 확률**<BR>
P(A|B) = $짝수일 확률 \over 3보다 클 확률$ = $P({4,6})\over P({4,5,6})$ = 2/3

**짝수가 나왔을때 그 수가 3보다 클 확률**<BR>
P(B|A) = $3보다 클 확률 \over 짝수일 확률$ = $P({4,6})\over P({2,4,6})$ = 2/3


```python
import matplotlib.pyplot as plt
import numpy as np

def main():
    # 그래프 크기 설정
    plt.figure(figsize=(5,5))
    
    # X와 Y의 빈 리스트
    X = []
    Y = []
    
    # n을 증가시킴
    N = 1000
    # random.random만 넣을시 0~1까지의 값만 생성한다.
    for i in range(N) :
        X.append(np.random.random())
        Y.append(np.random.random())
    X = np.array(X)
    Y = np.array(Y)
    X = X * 2 - 1 #[0, 1] --> [0, 2] -->  [-1, 1]로 바뀜!
    Y = Y * 2 - 1
    # 원점에서의 거리
    dist = np.sqrt(X**2 + Y**2)
    # 원점에서의 거리가 1보다 같거나 작은 것만 true 
    is_inside_circle = dist <= 1
    
    print("Estimated pi = %f" % (np.average(is_inside_circle) * 4))
    
    # 그래프 그리기
    plt.scatter(X, Y, c=is_inside_circle)
    plt.show()

if __name__ == "__main__":
    main()
```

### 베이즈 법칙

#### 빈도주의자 vs 베이즈주의자

"동전 하나를 던졌을 때 앞면이 나올 확률은 50%이다."

**빈도주의자** : 이 동전을 수천, 수만번 던졌을 때 그 중 앞면이 50%, 뒷면이 50% 나온다.

**베이즈주의자** : 동전 던지기의 결과가 앞면이 나올 것이라는 확신이 50%이다.


P(A|X) = $(P(X|A)P(A) \over P(X))$


#### 예제 : 암 검사 키트

암 A에 대한 테스트 키트가 있다.

임의의 사람이 이 암에 걸릴 확률은 1%이다. <BR>
즉, 전체 인구 중 암에 걸릴 사람은 1%이다.

이 암을 진단할 수 있는 키트가 있는데, 암에 걸린 사람은 99%의 확률로 양성 반응이 나오고, 걸리지 않은 사람은 1%의 확률로 양성 반응이 나온다.

키트 검사 결과, 양성 반응이 나왔다면, 암에 걸렸을 확률은?


암에 걸린 사건 : A<BR>
키트에서 양성반응이 나온 사건 : X

**키트에서 양성반응이 나왔을 때 암에 실제로 걸렸을 확률**<BR>
$P(A|X) =$ $P(X|A)P(A) \over P(X)$

**암에 걸렸을 때 키트에서 양성반응이 나올 확률** <BR>
P(X|A) = 0.99

**임의의 사람이 이 암에 걸릴 확률**<BR>
P(A) = 0.01

P(X) = P(X∩A) + P(X∩NOT A) <BR>
= P(X|A)P(A) + P(X|NOT A)P(NOT A) <BR>
    = 0.99 * 0.01 + 0.01 * 0.99 = 0.0198
  
P(A|X) = $P(X|A)P(A) \over P(X)$ <BR>
= $0.99 \times 0.01 \over 0.0198$ = 0.5

```python
# 베이즈 법칙
# sensitivity : 검사의 민감성 -> 실제로 해당 환자가 유방암을 가지고 있을 획률
# prior_prob : 총 인구를 기준으로 유방암을 가지고 있을 사전 확률
# false_alarm : 실제로 병을 가지고 있지 않은데도 유방암이라고 진단될 확률

def main():
    sensitivity = float(input())
    prior_prob = float(input())
    false_alarm = float(input())

    print("%.2lf%%" % (100 * mammogram_test(sensitivity, prior_prob, false_alarm)))

def mammogram_test(sensitivity, prior_prob, false_alarm):
    p_a1_b1 = sensitivity # p(A = 1 | B = 1), 실제로 암을 가지고 있을 확률

    p_b1 = prior_prob    # 유방암을 가지고 있을 확률

    p_b0 = 1 - p_b1    # 암을 가지고 있지 않을 확률

    p_a1_b0 = false_alarm # p(A = 1|B = 0) 암을 가지고 있지 않은데 암이 있다고 진단 받을 확률

    p_a1 = p_a1_b0 * p_b0 + p_a1_b1 * p_b1    # p(A = 1) = p(A = 1|B = 0)P(B = 0) + P(A = 1|B = 1)P(B = 1)

    p_b1_a1 = p_a1_b1 * p_b1/ p_a1 # p(B = 1|A = 1)

    return p_b1_a1

if __name__ == "__main__":
    main()
```

문제 : 사탕 10개를 뽑아서 빨강색 4개, 노랑색 5개, 초록색 1개를 뽑았다면 이 사탕은 어느 기계에서 뽑은 것일까?

-> X: 사탕 10개를 뽑아 그 결과를 관측한 사건<BR>
   A: 사탕 기계 A에서 사탕을 뽑은 사건<BR>
   B: 사탕 기계 B에서 사탕을 뽑은 사건
    
P(A|X)와 P(B|X)를 비교하면 어떤 것이 더 클 것인가?

**사탕을 뽑았을 때 A에서 10개를 뽑은 경우** <BR>
$P(A|X)$ = $P(X|A)P(A) \over P(X)$ <BR>

**사탕을 뽑았을 때 B에서 10개를 뽑은 경우** <BR>
$P(B|X)$ = $P(X|B)P(B) \over P(X)$
    

P(A|X) : P(B|X) 는 분모가 같으므로 <BR>
    = P(X|A)P(A) : P(X|B)P(X|B)를 비교하면 된다.
    
    
### 사전 확률

A기계보다 B기계가 조금 더 좋은 자리에 있어서 일반적으로 더 많이 팔린다.

P(A) = 0.4, P(B) = 0.6

$P(A|X)$ = $P(X|A)P(A) \over P(X)$에서 $P(A)$는 사전확률

### Likelihood(우도)

|비율|빨강생|노랑색|초록색|
|----|-----|------|------|
|A|2|2|1|
|B|1|1|1|

가정 : 사탕 기계가 매우 커서, 그 안에 있는 사탕 수의 비율은 몇 개를 꺼내도 일정하게 유지된다.

A에서 빨강 사탕을 꺼낼 확률 = $2 \over 5$ <br>
A에서 노랑 사탕을 꺼낼 확률 = $2 \over 5$ <br>
A에서 초록 사탕을 꺼낼 확률 = $1 \over 5$


A에서 빨강 사탕 4개를 꺼낼 확률 = $({2 \over 5})^4$ <br>
A에서 노랑 사탕 5개를 꺼낼 확률 = $({2 \over 5})^5$ <br>
A에서 초록 사탕 1개를 꺼낼 확률 = $({1 \over 5})^1$

A에서 빨강 사탕 4개,노랑 사탕 5개,A에서 초록 사탕
1개<br>
$P(X|A)$= $({2 \over 5})^4 \times ({2 \over 5})^5 \times ({1 \over 5})^1$ = $5.243 * (10)^{-5}$

B에서 빨강 사탕 4개,노랑 사탕 5개,A에서 초록 사탕 1개 <BR>
$P(X|B)$= $({1 \over 3})^4 \times ({1 \over 3})^5 \times ({1 \over 3})^1$ = $1.694 * (10)^{-5}$
    


그러나 앞서 풀이는 순서를 고려하지 않고 
처음 빨강 4개, 노랑 사탕 5개, 초록 사탕 1개인 경우만 계산하였으며

나올 수 있는 모든 경우의 수를 계산한 식은 다음과 같다.<br>
$P(X|A)$= $({2 \over 5})^4 \times ({2 \over 5})^5 \times ({1 \over 5})^1 \times C$

$P(X|B)$= $({1 \over 3})^4 \times ({1 \over 3})^5 \times ({1 \over 3})^1 \times C$

그러나 
$P(A|X) : P(B|X) =  P(X|A)P(A) : P(X|B)P(X|B)$이므로 위에 식의 C를 무시해도 된다.

따라서 $P(A|X) : P(B|X)$ <br> 
$=$ $P(X|A)P(A) \over P(X)$ $:$ $P(X|B)P(X|B) \over P(X)$ <br> 
$=$  $P(X|A)P(A) : P(X|B)P(B)$<BR>
= $5.243 * (10)^{-5} \times 0.4 : 1.694 * (10)^{-5} \times 0.6$ <br>
= 0.674 : 0.326

꺼낸 10개의 사탕은 A에서 나왔을 확률이 B에서 나왔을 확률보다 두 배 더 높다.


```python
import re
import math
import numpy as np

def main():
    M1 = {'r': 0.7, 'g': 0.2, 'b': 0.1} # M1 기계의 사탕 비율
    M2 = {'r': 0.3, 'g': 0.4, 'b': 0.3} # M2 기계의 사탕 비율
    
    test = {'r': 4, 'g': 3, 'b': 3}

    print(naive_bayes(M1, M2, test, 0.7, 0.3))

def naive_bayes(M1, M2, test, M1_prior, M2_prior):
    # 우도 구하기 
    M1_likelihood = M1['r'] ** test['r'] *M1['g']**test['g'] * M1['b']**test['b']
    M2_likelihood = M2['r'] ** test['r'] *M2['g']**test['g'] * M2['b']**test['b']
    # 사전 확률 곱하기
    M1_posterior = M1_likelihood * M1_prior
    M2_posterior = M2_likelihood * M2_prior
    # 표준화
    M1_normalized = M1_posterior / (M1_posterior + M2_posterior )
    M2_normalized = M2_posterior / (M1_posterior + M2_posterior )
    return [M1_normalized,M2_normalized]

if __name__ == "__main__":
    main()
```

### Bag of Words , 감정분석


오늘 나는 밥을 먹었다. 어제 나는 햄버거를 먹었다. <br>
-> **특수 문자 제거<br>**
오늘 나는 밥을 먹었다 어제 나는 햄버거를 먹었다<br>
-> **Tokenize<br>**(공백을 기준으로 자름)
오늘/나는/밥을/먹었다/어제/나는/햄버거를/먹었다.<br>


Bag of Words 모델에서 순서는 중요하지 않다.<br>
python dictionary를 이용하면 쉽다.<BR>
ex)
{"오늘" : 1, "나는" : 2, "먹었다":2, "햄버거를": 1, "밥을" : 1, "어제" : 1 }



```python
import re

special_chars_remover = re.compile("[^\w'|_]")

# 메인함수
def main():
    sentence = "Bag-of-Words 모델을 Python으로 직접 구현하겠습니다."
    bow = create_BOW(sentence)

    print(bow)


def create_BOW(sentence):
    # 빈 딕셔너리
    bow = {}
    # sentence 소문자로 변환
    sentence_lower = sentence.lower()
    # 특수 문자 제거
    sentence_without_special_characters = remove_special_characters(sentence_lower)
    # 공백마다 분리(token으로 구분)
    splitted_sentence = sentence_without_special_characters.split()
    # 빈 토큰이 나올 수 있으니 빈토큰과 아닌 것 구분
    splitted_sentence_filtered = [
        token
        for token in splitted_sentence
        if len(token) >= 1
    ]
    for token in splitted_sentence_filtered :
        if token not in bow :
            bow[token] = 1
        else :
            bow[token] += 1
    # 다른 방식으로도 가능하다.
    # for token in splitted_sentence_filtered:
        #bow.setdefault(token,0)
        #bow[token] += 1
    return bow


def remove_special_characters(sentence):
    return special_chars_remover.sub(' ', sentence)


if __name__ == "__main__":
    main()
```
